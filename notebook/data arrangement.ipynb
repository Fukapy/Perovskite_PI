```python
# import modules
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from process import *
import datetime
```

```python
# raw database
raw_DF = pd.read_csv("data/raw/Perovskite_database_content_all_data.csv")
```

```python
# Select X and y
DF_1 = pd.concat([raw_DF.iloc[:,13:261],raw_DF["JV_default_PCE"]],axis=1)
# add Original_index
DF_1["Original_index"]=DF_1.index
```

```python
DF_1.to_csv("data/temp/DF_1.csv",index = False)
DF_1
```

```python
# Delete rows with missing essential information
# essential information
essential_list=["Substrate_stack_sequence", "ETL_stack_sequence", "HTL_stack_sequence", "Backcontact_stack_sequence",
                "Perovskite_composition_a_ions", "Perovskite_composition_b_ions", "Perovskite_composition_c_ions",
               "JV_default_PCE"]
```

```python
miss_index = []
for i in range(len(DF_1)):
    for essential_list_count in range(len(essential_list)):
        if DF_1[essential_list[essential_list_count]][i]=="Unknown" \
        or DF_1[essential_list[essential_list_count]][i]=="none" \
        or pd.isna(DF_1[essential_list[essential_list_count]][i]):
            miss_index.append(i)
print("miss count:", len(miss_index))
print("miss line count:", len(set(miss_index)))
```

```python
DF_2 = DF_1.drop(DF_1.index[miss_index], inplace=False)
DF_2 = DF_2.reset_index(drop=True)
```

```python
DF_2.to_csv("data/temp/DF_2.csv",index = False)
DF_2
```

```python
# Converted to an array of str type data for easier processing
str_array = np.array(DF_2.astype(str))
# Replace non-ASCII-coded characters with ASCII codes whenever possible
for i in range(len(str_array)):
    for j in range(len(str_array[0])):
        str_array[i][j] = str_array[i][j].translate(str.maketrans\
                                                    ({"‑":"-","‐":"-","–":"-",
                                                      "’":"'","′":"'","´":"'",
                                                      "∣":"|","µ":"u","α":"a"}))
```

```python
# Delete rows containing non-ASCII characters and commas in cells
error_index = []
for i in range(len(str_array)):
    for j in range(len(str_array[0])):
        if str_array[i][j].isascii()==False or "," in str_array[i][j]:
            error_index.append(i)
print("error count:",len(error_index))
print("error line count:",len(set(error_index)))
```

```python
DF_3 = DF_2.drop(DF_2.index[error_index], inplace=False)
DF_3 = DF_3.reset_index(drop=True)
```

```python
DF_3.to_csv("data/temp/DF_3.csv",index = False)
DF_3
```

```python
# Sorting out what CBFV libraries are available and what are not
t1 = datetime.datetime.now()
mag_df_3 = cbfv_table("Perovskite_composition_long_form",DF_3,elem_prop="magpie")
t2 = datetime.datetime.now()
print("elapsed time:",t2-t1)
```

```python
cbfv_error_index = []
for i in range(len(mag_df_3)):
    mag_sum = sum(mag_df_3.iloc[i])
    if mag_sum == 0:
        cbfv_error_index.append(i)
print(len(cbfv_error_index))
```

```python
DF_4 = DF_3.drop(DF_3.index[cbfv_error_index], inplace=False)
DF_4 = DF_4.reset_index(drop=True)
```

```python
DF_4.to_csv("data/temp/DF_4.csv",index = False)
DF_4
```
